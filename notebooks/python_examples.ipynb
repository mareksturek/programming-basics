```python
# Law of large numbers

import numpy as np
from numpy.random import randn

N = 100000
counter = 0
for i in randn(N):
    if i > -1 and i < 1:
        counter = counter + 1
counter/N
```

```python
# Profit margin calculation after tax (20%)

import random

revenue = []
expenses = []
profit = []
profit_after_tax = []
profit_margin = []

for i in range(0,11):
    n = random.randint(9000,25000)
    revenue.append(n)
print(revenue)
    
for i in range(0,11):
    m = random.randint(3000,20000)
    expenses.append(m)
print(expenses)

for i in range(0, len(revenue)):
    profit.append(revenue[i] - expenses[i])
print(profit)

tax = [round(i * 0.2, 2) for i in profit]
print(tax)

for i in range(0, len(profit)):
    profit_after_tax.append(profit[i] - tax[i])
print(profit_after_tax)

for i in range(0, len(profit)):
    profit_margin.append(profit_after_tax[i] / revenue[i])

profit_margin = [round((i*100),2) for i in profit_margin]
print(profit_margin)
```

```python
# Breast Cancer dataset with Seaborn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
df_cancer = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'], ['target']))
```

```python
df_cancer.columns
```

```python
df_cancer.head()
```

```python
sns.pairplot(df_cancer, hue = 'target', vars = ['mean radius', 'mean texture', 'mean area', 'mean perimeter', 'mean smoothness'] )
```

```python
plt.figure(figsize = (20, 10)) 
sns.heatmap(df_cancer.corr(), annot = True)
```

```python
class_0_df = df_cancer[ df_cancer['target']==0 ]
class_1_df = df_cancer[ df_cancer['target']==1 ]
```

```python
plt.figure(figsize=(10, 7))
sns.distplot(class_0_df['mean radius'], bins = 25, color = 'blue')
sns.distplot(class_1_df['mean radius'], bins = 25, color = 'red')
plt.grid()
```
